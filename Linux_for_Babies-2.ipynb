## 1. `ls` ‚Äî See inside folder
Mouse analogy: **Open folder & view files**.

````python
ls
````

## 2. `cd` ‚Äî Move to folder
Mouse analogy: **Double-click folder**.

````python
cd
````

## 3. `pwd` ‚Äî Where am I?
Mouse analogy: **Address bar in Explorer/Finder**.

````python
pwd
````

## 4. `mkdir` ‚Äî New folder
Mouse analogy: **Right-click ‚Üí New Folder**.

````python
mkdir demo_folder
````

## 5. `touch` ‚Äî New file
Mouse analogy: **Right-click ‚Üí New File**.

````python
touch demo_folder/example.txt
````

## 6. `cp` ‚Äî Copy
Mouse analogy: **Copy + Paste**.

````python
cp demo_folder/example.txt demo_folder/example_copy.txt
````

## 7. `mv` ‚Äî Move/Rename
Mouse analogy: **Drag & drop or rename**.

````python
mv demo_folder/example_copy.txt demo_folder/renamed.txt
````

## 8. `rm` ‚Äî Delete
Mouse analogy: **Delete permanently (‚ö†Ô∏è no trash)**.

````python
rm demo_folder/renamed.txt
````

## 9. `cat` ‚Äî Quick preview
Mouse analogy: **Double-click file**.

````python
cat demo_folder/example.txt
````

## 10. `grep` ‚Äî Search text
Mouse analogy: **CTRL+F**.

````python
grep 'hello' demo_folder/example.txt | echo 'not found'
````

# Cheat Sheet

| Action | Linux Command | Mouse Analogy |
|--------|---------------|----------------|
| See files | `ls` | Open folder |

| Go inside | `cd folder` | Double-click folder |
| New folder | `mkdir` | Right-click ‚Üí New folder |
| New file | `touch` | Right-click ‚Üí New file |
| Copy | `cp` | Copy + Paste |
| Move / Rename | `mv` | Drag / Rename |
| Delete | `rm` | Delete permanently |
| Read file | `cat` | Quick open |
| Search | `grep` | CTRL+F |
